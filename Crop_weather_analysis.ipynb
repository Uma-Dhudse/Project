```python
import pandas as pd

rain = pd.read_csv("C:\\my_project\\data\\rainfall.csv")
rain.head()
```

```python
crop = pd.read_csv("C:\\my_project\\data\\crop.csv")
crop.head()

```

```python
# Clean crop dataset
crop_clean = crop.rename(columns={
    'State_Name': 'state',
    'District_Name': 'district',
    'Crop_Year': 'year',
    'Season': 'season',
    'Crop': 'crop',
    'Area': 'area',
    'Production': 'production'
})

crop_clean.head()
```

```python
temp = pd.read_csv("C:\\my_project\\data\\temperatures.csv")
temp.head()
```

```python
# Clean temperature dataset
temp_clean = temp[['YEAR', 'ANNUAL']].rename(columns={
    'YEAR': 'year',
    'ANNUAL': 'annual_temperature'
})

temp_clean.head()
```

```python
rain.head()
```

```python
# Clean rainfall dataset (national-level)
rain_clean = rain[['year', 'annual_rainfall']]

rain_clean.head()
```

```python
# Merge crop + rainfall
merged = crop_clean.merge(rain_clean, on='year', how='left')

# Merge with temperature
merged = merged.merge(temp_clean, on='year', how='left')

merged.head()
```

```python
merged.info()
```

```python
merged.describe()
```

```python
merged.isnull().sum()
```

```python
# Fill missing production with 0
merged['production'] = merged['production'].fillna(0)

# Fill missing rainfall with average rainfallssssssssss
avg_rainfall = merged['annual_rainfall'].mean()
merged['annual_rainfall'] = merged['annual_rainfall'].fillna(avg_rainfall)

# Check again
merged.isnull().sum()
```

```python
merged[['production', 'annual_rainfall']].corr()
```

```python
merged[['production', 'annual_temperature']].corr()
```

```python
yearly = merged.groupby('year').agg({
    'production': 'sum',
    'annual_rainfall': 'mean',
    'annual_temperature': 'mean'
}).reset_index()

yearly.head()
```

```python
top_states = merged.groupby('state')['production'].sum().sort_values(ascending=False).head(5)
top_states
```

```python
top_crops = merged.groupby('crop')['production'].sum().sort_values(ascending=False).head(5)
top_crops
```
